```python
# IMPORT PARKAGES

import pandas as pd
import numpy as np
```

```python
# Set seed for reproducibility (consistent with MLflow principles)
np.random.seed(42)

# 1. Define Dataset Size
n_samples = 10000

# 2. Generate Base Features (Alternative Data)
data = {
    'Applicant_ID': range(1001, 1001 + n_samples),
    
    # Stability: How long have they had the phone number? (Months)
    'Sim_Tenure_Months': np.random.randint(1, 120, n_samples),
    
    # Consistency: Airtime/Data recharge patterns (0.0 = chaotic, 1.0 = robotic regularity)
    'Airtime_Consistency_Score': np.random.uniform(0.1, 1.0, n_samples),
    
    # Velocity: Average monthly mobile wallet inflows (Estimated in Rands)
    'Avg_Wallet_Inflow': np.random.normal(4500, 1500, n_samples).clip(500, 15000),
    
    # Discipline: Days between electricity recharges (Lower is better/more consistent)
    'Electricity_Payment_Gap_Days': np.random.randint(5, 45, n_samples),
    
    # Digital Savvy: Data usage vs Voice usage ratio
    'Data_Voice_Ratio': np.random.uniform(0.5, 5.0, n_samples)
}

df = pd.DataFrame(data)

# 3. Create a "Hidden Logic" for the Model to learn (The Probability of Default)
# We calculate a 'Risk Score' based on our pillars: Stability, Consistency, and Velocity
# Lower Score = Better Payer
risk_score = (
    (120 - df['Sim_Tenure_Months']) * 0.2 + 
    (1.0 - df['Airtime_Consistency_Score']) * 30 + 
    (df['Electricity_Payment_Gap_Days']) * 0.5 - 
    (df['Avg_Wallet_Inflow'] / 1000) * 2
)

# Convert Risk Score to a Probability (0 to 1) using a Sigmoid function
prob_default = 1 / (1 + np.exp(-(risk_score - np.mean(risk_score)) / 10))

# 4. Generate the Target Variable (Default_Status)
# 1 = Default (Bad), 0 = Paid on time (Good)
df['Default_Status'] = np.where(np.random.rand(n_samples) < prob_default, 1, 0)

# 5. Save the Dataset
df.to_csv('financial_inclusion_data.csv', index=False)

print("✅ Success! 'financial_inclusion_data.csv' has been generated with 10,000 rows.")
print(df.head())
```
